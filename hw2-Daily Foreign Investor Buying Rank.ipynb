```python
import requests as re
from pyquery import PyQuery as pq
```

```python
date=input("輸入日期 Ex:0310\n")
url='https://www.twse.com.tw/fund/TWT38U?response=html&date=2020{}'.format(date)
res=re.get(url)
doc=pq(res.text)
```

```python
list=[]
for i in range(31):
    for eachdoc in doc('body > div > table > tbody > tr:nth-child({})'.format(i)).items():
        dict={}
        dict["代號"]=eachdoc('td:nth-child(2)').text()
        dict["名稱"]=eachdoc('td:nth-child(3)').text()
        net=eachdoc('td:nth-child(6)').text()
        net=net.replace(',','')
        net=float(net)/1000
        net=str(round(net))
        dict["淨買賣"]=net+'張'
        list.append(dict)
print(list)
```

```python
str1=''
for i in list:
    for key,value in i.items():
        str1=str1+key+':'+str(value)+'\n'
    str1=str1+'\n'
print(str1)
```
