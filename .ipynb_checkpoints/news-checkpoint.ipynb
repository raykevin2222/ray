```python
#鉅亨網新聞 台灣
import pandas as pd
import requests
from pyquery import PyQuery as pq
res=requests.get("https://news.cnyes.com/news/cat/tw_stock")
html=pq(res.text)
newsTaiwan=[]
for i in html('._1xc2').items():
    newsTaiwan.append(i.text())
#print(newsTaiwan)

newsDomestic=""
newsTw=[]
n=0
for i in newsTaiwan:
    newsDomestic+=i
    n+=1
    if n==2:
        n=0
        newsTw.append(newsDomestic)
        newsDomestic=""
print(newsTw)

print('-----------------------------------------------------')
#鉅亨網新聞 國際
from pyquery import PyQuery as pq
res=requests.get("https://news.cnyes.com/news/cat/wd_stock")
html=pq(res.text)
newsGlobal=[]
for i in html('._1xc2').items():
    newsGlobal.append(i.text())

newsGlobal1=""
newsInt=[]
n=0
for i in newsGlobal:
    newsGlobal1+=i
    n+=1
    if n==2:
        n=0
        newsInt.append(newsGlobal1)
        newsGlobal1=""
print(newsInt)
    
```

```python
str1 = ''

for i in newsTw:
    str1 = str1+i+'\n'

str1 = str1+'-------------------------------------\n'

for i in newsInt:
    str1 = str1+i+'\n'

str1 = str1+'-------------------------------------\n'

print(str1)
```
